predicting whether this new customer, will buy this products or not

## Logistic Regression


```python
import pandas as pd
```

```python
data=pd.read_csv("DigitalAd.csv")
```

```python
data.head(5)
```

```python
data.shape
```

```python
x= data.iloc[:,:-1].values
```

```python
x
```

```python
y=data.iloc[:,-1].values
```

```python
y
```

## Spliting the dataset Train and Test

```python
!pip install scikit_learn
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test =train_test_split(x,y,test_size = 0.25,random_state=0)
```

## Feature Scaling

Since both the features have different scale, there is a chance that higher magnitude. This will impact the performance of the machine learning algoritham and obviously,we do not want our algoritham to be biased towards one feature.

```python
from sklearn.preprocessing import StandardScaler
sc =StandardScaler()
x_train =sc.fit_transform(x_train)
x_test = sc.transform(x_test)
print(x_train)
```

## Training

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(x_train,y_train)
```

## Predicting , wheathr new customer with Age &salary will Buy or Not

```python
age =int(input("Enter New customer's Age"))
sal= int(input("Enter New customer's Salary"))
newCust =[[age,sal]]
result = model.predict(sc.transform(newCust))
print(result)
if result ==1:
    print("customer will Buy")
else:
    print("customet won't Buy")
```

## Prediction for all Test Data

```python
y_pred = model.predict(x_test)
```

```python
y_pred 
```

```python
from sklearn.metrics import accuracy_score
print("Accuracy of the model : {0}%".format(accuracy_score(y_test,y_pred)*100))
```
